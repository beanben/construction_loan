```python
# prepare the dataframe
df = budget.set_project_budget_from_csv('/Users/benoitfesquet/Dev/construction_loan/data/input_data.csv')
 
df
```

```python
cashflow_df = budget.spread_costs_evenly(df)

cashflow_df
```

```python
cashflow_df_formatted = cashflow_df.copy()
cashflow_df_formatted.index = cashflow_df_formatted.index.date

def style_df(styler):
    styler.set_table_styles([{
        'selector': 'th',
        'props': [('background-color', '#f7f7f9'), 
                  ('border-left', '1px solid black'),
                  ('border-right', '1px solid black'),
                  ('font-weight', 'bold')]
    }])

    
    # Apply number formatting, rounding to whole numbers
    format_spec = {col: '{:,.0f}' for col in styler.data.select_dtypes(include=['float64', 'int']).columns}
    
    styler.format(format_spec)
    
    return styler


cashflow_df_formatted = cashflow_df_formatted.style.pipe(style_df)


cashflow_df_formatted
```
